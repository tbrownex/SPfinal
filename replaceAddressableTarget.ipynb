```python
import pandas as pd
import json

from utils import putRequest
from config import getBaseURL
from getAddresses import getAddresses
```

```python
def getCampaigns():
    df = pd.read_csv('/home/tbrownex/data/SteveRoss/Oct2022/campaign_id_map.csv')
    df.columns=['campaignID', 'id', 'processFlag']
    df=df[df['processFlag']=='Y']
    df['campaignID'] = df['campaignID'].astype(int)
    return df
```

```python
def getID(x):
    resp = getAddresses(x)
    return resp['campaign_addresses'][0]['id']
```

```python
def getBaseURL():
    return "https://app.simpli.fi/api/"
```

```python
orgID = 334178    # Bixy
path = '/home/tbrownex/data/SteveRoss/Oct2022/'

base = getBaseURL()
```

```python
campaigns = getCampaigns()
```

```python
campaigns['delAddr'] = campaigns['campaignID'].apply(lambda x: getID(x))
```

```python
for row in campaigns.iterrows():
    campaignID = row[1]['campaignID']
    delID = int(row[1]['delAddr'])
    url = base + "campaigns/"+str(campaignID)+'/campaign_addresses/change'
    d = {}
    d["delete"] = [delID]
    payload = {}
    payload["campaign_first_party_segments"] = d
    payloadString = json.dumps(payload)
    resp=putRequest(url, payloadString)
```
